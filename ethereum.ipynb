```python
import pandas as pd
import yfinance as yf
from datetime import datetime
from datetime import timedelta
import plotly.graph_objects as go
from fbprophet import Prophet
from fbprophet.plot import plot_plotly, plot_components_plotly
import warnings 
warnings.filterwarnings('ignore')
pd.options.display.float_format = '{:,.2f}'.format
```

```python
hoje = datetime.today().strftime('%Y-%m-%d')
```

```python
data_ini = '2016-01-01'
df_eth = yf.download('ETH-USD', data_ini, hoje)
df_eth.tail()
```

```python
df_eth.reset_index(inplace=True)
```

```python
df_eth
```

```python
df = df_eth[['Date', 'Adj Close']]
df.rename(columns = {'Date': 'ds', 'Adj Close': 'y'}, inplace=True)
```

```python
df
```

```python
#Gráfico preço de fechamento
fig = go.Figure()
fig.add_trace(go.Scatter(x=df['ds'], y = df['y']))
```

```python
#Construção do Modelo
model = Prophet(seasonality_mode='multiplicative', daily_seasonality=True)
model.fit(df)
```

```python
#Exibe o modelo preditivo
df_futuro = model.make_future_dataframe(periods=60)
df_futuro.tail()
```

```python
previsao = model.predict(df_futuro)
```

```python
plot_plotly(model, previsao)
```

```python
plot_components_plotly(model, previsao)
```
